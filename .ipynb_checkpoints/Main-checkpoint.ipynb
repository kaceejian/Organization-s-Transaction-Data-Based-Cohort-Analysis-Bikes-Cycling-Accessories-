# Further Data Cleaning - Checking and working with missing value

## 1-A. Numerical Variables - Transaction DF Cleaning:

```python
# Inspect missing values in the dataset
print(transaction_df.isnull().values.sum())
```

```python
# Replace the ' 's with NaN
transaction_df = transaction_df.replace(" ",np.NaN)
```

```python
# Impute the missing values with mean imputation
transaction_df = transaction_df.fillna(transaction_df.mean())
```

```python
# Count the number of NaNs in the dataset (verify again after filling in the missing values)

print(transaction_df.isnull().values.sum())
```

## 1-B. Numerical Variables - Demographic DF Cleaning:

```python
# Inspect missing values in the dataset
print(demographic_df.isnull().values.sum())
```

```python
# Replace the ' 's with NaN
demographic_df = transaction_df.replace(" ",np.NaN)
```

```python
# Impute the missing values with mean imputation
demographic_df = transaction_df.fillna(transaction_df.mean())
```

```python
# Count the number of NaNs in the dataset (verify again after filling in the missing values)

print(demographic_df.isnull().values.sum())
```

## 2-A. Categorical Variables - Transaction DF Cleaning:

```python
print(transaction_df.info())
```

```python
# 以上check info 显示的： 是不是这次再检查之后，剩下的还有 Null 的就都是 Categorical 的 missing values 了？
```

```python
print(transaction_df.info())
for col in transaction_df.columns:
    
    # Check if the column is of object type
    if transaction_df[col].dtypes == 'object':
       
        # If the column is "object" type, then input with the most frequent value (find the most frequent value by-
        #- finding the value_counts, and the top row, which is 'index[0], is the most frequent value we want to find. 
        transaction_df[col] = transaction_df[col].fillna(transaction_df[col].value_counts().index[0])

# Count the number of NaNs in the dataset and print the counts to verify if it is '0' now. (which means 0 null value.)
print(transaction_df.isnull().values.sum())
```

## 2-B. Categorical Variables - Demographic DF Cleaning:

```python
print(demographic_df.info())
```

```python
print(demographic_df.info())
for col in demographic_df.columns:
    
    # Check if the column is of object type
    if demographic_df[col].dtypes == 'object':
       
        # If the column is "object" type, then input with the most frequent value (find the most frequent value by-
        #- finding the value_counts, and the top row, which is 'index[0], is the most frequent value we want to find. 
        demographic_df[col] = demographic_df[col].fillna(demographic_df[col].value_counts().index[0])

# Count the number of NaNs in the dataset and print the counts to verify if it is '0' now. (which means 0 null value.)
print(demographic_df.isnull().values.sum())
```
